# Replacing blanks with 0 and no total charges are recorded

```python
df['TotalCharges'] = df['TotalCharges'].replace(" ", "0")
df['TotalCharges'] = df['TotalCharges'].astype("float")
```

# Converted 0 and 1 value of senior citizen to yes or no

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"

df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
```

```python
df.head(30)
```

# How many customers and churn in and churn out

```python
ax = sns.countplot(x = df['Churn'], data = df)
ax.bar_label(ax.containers[0])
plt.title("Count Of Churn")
plt.show()
```

# To look in percentage we are using pie chart

```python
gb = df.groupby('Churn').agg({'Churn':'count'})
plt.pie(gb['Churn'], labels=gb.index, autopct='%1.2f%%')
plt.title("Percentage Of Churn")
plt.show()
gb
```

# from the given pie chart we can conclude that 26.54% of our customers have churned out.
# now lets explore the reason behind it

```python
plt.figure(figsize= (4,4))
plt.title("Churn by gender")
sns.countplot(x = 'gender', data = df, hue = 'Churn')
plt.show()
```

```python
plt.figure(figsize= (4,4))
plt.title("Churn by SeniorCitizen")
sns.countplot(x = 'SeniorCitizen', data = df, hue = 'Churn')
plt.show()
```

```python
# Calculate % per SeniorCitizen group
total_counts = df.groupby('SeniorCitizen')['Churn'].value_counts(normalize=True).unstack() * 100

# Create plot
fig, ax = plt.subplots(figsize=(5,5))

# Plot stacked bar chart
total_counts.plot(kind='bar', stacked=True, ax=ax, color=['#1f77b4','#ff7f0e'])

# Add percentage labels
for p in ax.patches:
    height = p.get_height()
    x, y = p.get_xy()
    ax.text(
        x + p.get_width()/2,
        y + height/2,
        f'{height:.1f}%',
        ha='center', va='center', color='white', fontweight='bold'
    )

# Formatting
plt.title('Churn by Senior Citizen (Stacked % per Group)')
plt.xlabel('SeniorCitizen')
plt.ylabel('Percentage')
plt.xticks(rotation=0)
plt.legend(title='Churn', bbox_to_anchor = (0.9,0.9))
plt.show()
```

# People who have used our services for a long time have stayed and people who have used our services #1 or 2 months have churned

# People who have month to month contract are likely to churn then from those who have 1 or 2 years of contract

```python
plt.figure(figsize= (4,4))
ax = sns.countplot(x = 'Contract', data = df, hue='Churn')
ax.bar_label(ax.containers[0])
plt.title("Count of customers by contract")
plt.show()
```

```python
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Create subplots
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15,10))
axes = axes.flatten()  # flatten 2D axes to 1D for easy indexing

for i, col in enumerate(cols):
    sns.countplot(x=col, data=df, hue=col, legend=False, palette='Set2', ax=axes[i],)
    axes[i].set_title(f"Count Plot of {col}")
    axes[i].set_xlabel("")
    axes[i].set_ylabel("Count")

plt.tight_layout()
plt.show()
```

# Most customers have PhoneService, with fiber optic being the most common internet type.
# However, the majority lack extra services like OnlineSecurity, OnlineBackup, DeviceProtection, and TechSupport.
# Streaming services (StreamingTV and StreamingMovies) are more evenly split between users who opt in and those who don’t.
# A small portion of customers have no internet service at all.
